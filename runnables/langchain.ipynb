```python
import random
class NakliLLM :
    def __init__(self):
        print("NakliLLM initialized")
    def predict(self, input_text):
        response_list=[
            "NakliLLM response 1",
            "NakliLLM response 2",
            "NakliLLM response 3",
            "NakliLLM response 4",
            "NakliLLM response 5"
        ]
        return {"response": random.choice(response_list)}
    
 
```

```python
class NakliPromptTemplate :
    def __init__(self, template,input_variables):
        self.template = template
        self.input_variables = input_variables
    def format(self, input_dict):
        return self.template.format(**input_dict)
```

```python
template  = NakliPromptTemplate(
    template= "Write a {length} poem about {topic}",
    input_variables= ["topic", "length"]
)
```

```python
prompt = template.format({"topic": "Nature","length": "short"})
```

```python
llm = NakliLLM()
```

```python
llm.predict(prompt)
```

```python
class NakliLLMChain :
    def __init__(self, llm, prompt_template):
        self.llm = llm
        self.prompt_template = prompt_template
    def run(self, input_dict):
        final_prompt = self.prompt_template.format(input_dict)
        result = self.llm.predict(final_prompt)
        return result['response']
```

```python
chain = NakliLLMChain(llm=llm, prompt_template=template)
```

```python
chain.run({"topic": "Technology", "length": "long"})
```
